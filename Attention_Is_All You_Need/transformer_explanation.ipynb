```python
import torch
import torch.nn as nn
import pandas as pd
import altair as alt
import matplotlib.pyplot as plt
import numpy as np
```

Mask Visualization

```python
def subsequent_mask(size):
    attn_shape = (1, size, size)
    mask = torch.triu(torch.ones(attn_shape), diagonal=1).type(torch.uint8)
    return mask == 0

mask = subsequent_mask(5).numpy()[0]
plt.imshow(mask, cmap='viridis')
plt.title("Subsequent Mask")
plt.savefig("mask.png")
```

Padding Example

```python
sentences = [[1, 2, 0], [1, 0, 0]]  # "Hello world", "Hello"
plt.imshow(sentences, cmap='gray')
plt.title("Padded Sentences")
plt.savefig("padding.png")
```

Probability Distribution

```python
scores = torch.tensor([2.3, 1.1, 0.5])
probs = torch.softmax(scores, dim=0)
plt.bar(["hola", "mundo", "adios"], probs.numpy())
plt.title("Word Probabilities")
plt.savefig("probs.png")
```

Attention Heatmap

```python
attention = np.random.rand(3, 3)  # Simulated attention weights
plt.imshow(attention, cmap='hot')
plt.title("Attention Weights")
plt.savefig("attention.png")
```

Residual Connection Demo

```python
x = np.array([1, 2, 3])
sublayer_output = np.array([0.1, 0.2, 0.3])
output = x + sublayer_output
plt.plot(x, label="Input")
plt.plot(output, label="Output with Residual")
plt.legend()
plt.savefig("residual.png")
```

Multi-Head Attention

```python
head1 = np.random.rand(3, 3)  # Syntax focus
head2 = np.random.rand(3, 3)  # Meaning focus
fig, axs = plt.subplots(1, 2)
axs[0].imshow(head1, cmap='hot')
axs[1].imshow(head2, cmap='hot')
plt.savefig("multihead.png")
```
