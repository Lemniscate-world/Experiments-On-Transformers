# Mask Visualization

```python
def subsequent_mask(size):
    attn_shape = (1, size, size)
    mask = torch.triu(torch.ones(attn_shape), diagonal=1).type(torch.uint8)
    return mask == 0

mask = subsequent_mask(5).numpy()[0]
plt.imshow(mask, cmap='viridis')
plt.title("Subsequent Mask")
plt.savefig("mask.png")
```

# Scaled Dot-Product Demo

```python
Q = np.array([[1, 2]])  # Query: "sleeps"
K = np.array([[0, 1], [2, 1], [1, 0]])  # Keys: "The", "cat", "sleeps"
dk = 2
scores = Q @ K.T  # Dot product
scaled_scores = scores / np.sqrt(dk)
weights = np.exp(scaled_scores) / np.sum(np.exp(scaled_scores))  # Softmax

plt.bar(['The', 'cat', 'sleeps'], weights[0], color='lightcoral')
plt.title('Scaled Dot-Product Attention Weights')
plt.savefig('scaled_attention.png')
plt.show()
```

```python
# Dot-Product vs. Additive Attention
Q = np.array([[1, 2]])  # Query: "sleeps"
K = np.array([[0, 1], [2, 1], [1, 0]])  # Keys
V = np.array([[0.1, 0.2], [0.5, 0.6], [0.3, 0.4]])  # Values

# Dot-Product
dk = 2
scores_dot = (Q @ K.T) / np.sqrt(dk)
weights_dot = np.exp(scores_dot) / np.sum(np.exp(scores_dot))

# Additive (simplified)
scores_add = np.sum(Q + K, axis=1)  # Fake neural network: just sum
weights_add = np.exp(scores_add) / np.sum(np.exp(scores_add))

fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].bar(['The', 'cat', 'sleeps'], weights_dot[0], color='lightblue')
axs[0].set_title('Dot-Product Attention')
axs[1].bar(['The', 'cat', 'sleeps'], weights_add, color='lightgreen')
axs[1].set_title('Additive Attention')
plt.savefig('attention_comparison.png')
plt.show()
```
